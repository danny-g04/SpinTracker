```python
#Used to show number of each stroke after a session
plt.bar(["Died", "Survived"], [549,342], color = ["red", "green"])
plt.title("Titanic survival counts")
plt.xlabel("Outcome")
plt.ylabel("Number of passengers")
plt.show() #shows chart (vscode does it auto)
```

```python
#Used to visualize confidence scores dist
plt.hist(df["Age"], bins = 20, rwidth=.8) #row width
plt.title("Ages of Passengers")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
#scatterplot for visualizing body landmark coordinates
plt.scatter(df["Age"], df["Fare"], alpha =0.5) #alpha is transparentetcy
plt.title("Age vs fare paid")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
